# Nulclinas
Las nulclinas son una forma de observar graficamente el comportamiento de un
sistema de dos ecuaciones, que podemos usar para encontrar y analizar sus
estados estables.
En el caso de un sistema de ecuaciones que describe los cambios de $x$ en
funcion de $y$ y viceversa, tendremos la nulclina de $x$ cuando
$\frac{dx}{dt} = 0$ y la de $y$ cuando $\frac{dy}{dt} = 0$.

Tomaremos como ejemplo un circuito donde $x$ y $y$ se regulan mutuamente de
forma positiva. La ecuacion que describe a $x$ en funcion de $y$:
\begin{align}
\frac{dx}{dt} = \beta_x\frac{y^n}{k^n + y^n} - \gamma_x x = 0
\end{align}
El estado estable de $x$, $x_{st}$, es: 
\begin{align}
x_{st} = \frac{\beta_x y^n}{\gamma_x (k^n + y^n)}
\end{align}
El estado estable de $y$ esta descrito por la misma ecuacion en funcion de
$x$:
\begin{align}
y_{st} = \frac{\beta_y x^n}{\gamma_y (k^n + x^n)}
\end{align}

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.integrate import odeint

# vector de x y y
y = np.linspace(0, 50, 1000)
beta_x = 1
gamma = 1/30
k = 5
n = 2
# estados estables, son la misma ecuacion, podemos simplemente invertir al graficar
x_act = (beta_x * y**n) / (gamma*(k**n + y**n))
plt.plot(y,x_act, 'b-', x_act, y, 'r-')
plt.margins(x=0.05, y=0.05)
```

Cuando $x$ y $y$ se regulan mutuamente de forma negativa, tenemos:
\begin{align}
\frac{dx}{dt} = \beta_x\frac{1}{1 + (\frac{y}{k})^n} - \gamma_x x = 0
\end{align}
Encontramos el estado estable:
\begin{align}
x_{st} = \frac{\beta_x}{\gamma_x (1 + \frac{y}{k}^n)}
\end{align}
De la misma forma para $y$:
\begin{align}
y_{st} = \frac{\beta_y}{\gamma_y (1 + \frac{x}{k}^n)}
\end{align}

```python
y = np.linspace(0, 50, 1000)
x_rep = beta_x / (gamma*(1 + (y/k)**n))
plt.xlabel('', fontsize=20)
plt.plot(y,x_rep, 'b-', x_rep, y, 'r-')
plt.xlabel('', fontsize=20)
```
